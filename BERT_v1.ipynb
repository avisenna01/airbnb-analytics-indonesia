```python
from textblob import TextBlob
import plotly.express as px

#Create a function to get the subjectivity
def subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

#Create a function to get the polarity
def polarity(text):
    return TextBlob(text).sentiment.polarity

#Create two new columns
df['Subjectivity'] = df['clean_reviews_v2'].apply(subjectivity)
df['Polarity'] = df['clean_reviews_v2'].apply(polarity)
#################################################################################
#Create a function to compute the negative, neutral and positive analysis
def getAnalysis(score):
    if score <0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'

df['Analysis'] = df['Polarity'].apply(getAnalysis)
#################################################################################
```

```python
# Create an empty dataframe
topics_by_group = pd.DataFrame()

# Loop through each neighbourhood
for i in city_list:
  for j in range_list:
    # A data frame only contains reviews from neighbourhood i
    group = df_negative_reviews[(df_negative_reviews['city'] == i) & (df_negative_reviews['range_period'] == j)]
    if (len(group)>0):
      # Reset index, otherwise there will be error from BERTopic model
      group.reset_index(drop=True, inplace=True)
      try:
        embeddings = sentence_model.encode(group['clean_reviews_v2'], show_progress_bar=False)
        # Train the topic model
        topics, probs = topic_model.fit_transform(group['clean_reviews_v2'], embeddings)
        # Get the topic list for the neighbourbhood i
        topics_df = topic_model.get_topic_info()
        # Create a column with the city name
        topics_df['city'] = i
        # Create a column with the range period name
        topics_df['range_period'] = j

        umap_embeddings = topic_model.umap_model.transform(embeddings)
        if -1 in topics and len(topics) == 1:
          indices = [index for index, topic in enumerate(topics)]
          if len(indices) != 0:
            X = umap_embeddings[np.array(indices)]
            labels = [topic for index, topic in enumerate(topics)]
          else:
            topics_df['silhouette_score'] = None
            topics_by_group = pd.concat((topics_by_group, topics_df))
            continue
        else:
          indices = [index for index, topic in enumerate(topics) if topic != -1]
          if len(indices) != 0:
            X = umap_embeddings[np.array(indices)]
            labels = [topic for index, topic in enumerate(topics) if topic != -1]
          else:
            topics_df['silhouette_score'] = None
            topics_by_group = pd.concat((topics_by_group, topics_df))
            continue

        # Create a column with the silhouette score
        topics_df['silhouette_score'] = silhouette_score(X, labels)

        # Append the topic list to the dataframe
        topics_by_group = pd.concat((topics_by_group, topics_df))
      except ValueError:
        pass
    else:
      continue

topics_by_group
```
